```python
import pandas as pd
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
adult = fetch_ucirepo(id=2) 
```

```python
# data (as pandas dataframes) 
X = adult.data.features 
df1= pd.DataFrame(X)
df1
```

```python
df1.fillna(method='ffill',inplace=True)
df1
```

```python
print(df1.isnull().sum())    # to make sure missing values are done 
```

```python
#drop useless columns 
df1.drop(['fnlwgt','education'],axis=1,inplace=True)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for column in df1.columns:
    if df1[column].dtype == 'object':  # Assuming 'object' type for categorical data
        df1[column] = LE.fit_transform(df1[column])
df1

```

```python
y = adult.data.targets 
df2=pd.DataFrame(y)
```

```python
df2['<=50K'] = (df2['income'] == '<=50K').astype(int)
df2['>50K'] = (df2['income'] == '>50K').astype(int)
df2
```

```python
df2.drop(['income','<=50K'],axis=1,inplace=True)
df2
```

```python
from sklearn.model_selection import train_test_split
df1_train , df1_test , df2_train , df2_test = train_test_split (df1 ,df2 , test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(df1_train , df2_train )
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df2_pred=nb.predict(df1_test)
print(confusion_matrix(df2_test,df2_pred))
print(accuracy_score(df2_test,df2_pred)*100)

# computing Sensitivity and Specificity
tn, fp, fn, tp = confusion_matrix(df2_test,df2_pred).ravel()  # ravel() flattens the 2x2 matrix into a 1D array

sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
print ("sensitivity = " ,sensitivity)
print ("specificity = " ,specificity)
```

```python
#Compute the posterior probability of making over 50K a year
#1st compute the prevalence 
prevalence = df2.mean() 


probability_exceeds50 = (sensitivity * prevalence) + ((1 - specificity) * (1 - prevalence))
posterior_probability = (sensitivity * prevalence) / probability_exceeds50
print ( "posterior_probability of" , posterior_probability)
```
